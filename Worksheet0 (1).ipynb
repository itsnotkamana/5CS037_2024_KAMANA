```python
#Task 4. Analyze Temperature Patterns by Time of Day:

import matplotlib.pyplot as plt

temperatures = [
    8.2, 17.4, 14.1, 7.9, 18.0, 13.5, 9.0, 17.8, 13.0, 8.5,
    16.5, 12.9, 7.7, 17.2, 13.3, 8.4, 16.7, 14.0, 9.5, 18.3, 
    13.4, 8.1, 17.9, 14.2, 7.6, 17.0, 12.8, 8.0, 16.8, 13.7, 
    7.8, 17.5, 13.6, 8.7, 17.1, 13.8, 9.2, 18.1, 13.9, 8.3, 
    16.4, 12.7, 8.9, 18.2, 13.1, 7.8, 16.6, 12.5
]

night_temps = []
day_temps = []
evening_temps = []

for i, temp in enumerate(temperatures):
    if i % 24 < 8:  
        night_temps.append(temp)
    elif i % 24 < 16:  
        evening_temps.append(temp)
    else: 
        day_temps.append(temp)

avg_day_temp = sum(day_temps) / len(day_temps)
print(f"Average day-time temperature: {avg_day_temp:.2f}Â°C")

days = list(range(1, len(day_temps) + 1))
plt.plot(days, day_temps, marker='o', label="Daytime Temperatures", color='orange')
plt.xlabel("Day")
plt.ylabel("Temperature (Â°C)")
plt.title("Day vs. Temperature")
plt.legend()
plt.grid(True)
plt.show()

```

```python
#8.1.1 Exercise - Recursion:Task 1 - Sum of Nested Lists:
def sum_nested_list(nested_list):
    
    total = 0  

    for element in nested_list:
        if isinstance(element, list):  
            total += sum_nested_list(element)
        else:
            total += element
    return total  

nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
result = sum_nested_list(nested_list)
print(f"The total sum of {nested_list} is: {result}")

nested_list2 = [1, [2, 3], [4, [5]]]
result2 = sum_nested_list(nested_list2)
print(f"The total sum of {nested_list2} is: {result2}")
```

```python
#Task 2 - Generate All Permutations of a String:

def generate_permutations(s):
    if len(s) <= 1:
        return [s]
    permutations = []
    for i in range(len(s)):

        char = s[i]

        remaining = s[:i] + s[i+1:]

        for perm in generate_permutations(remaining):
            permutations.append(char + perm)
            
    return sorted(list(set(permutations))) 
    
print(generate_permutations("abc"))

print(generate_permutations("aab"))




    
```

```python
#Task 3 - Directory Size Calculation:
def calculate_directory_size(directory):
    total_size = 0

    for item, value in directory.items():
        if isinstance(value, dict):
            total_size += calculate_directory_size(value)
        else:
            # If it's not a dictionary, it's a file, so add its size to the total
            total_size += value

    return total_size  # Return the total size

directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

print(calculate_directory_size(directory_structure))


```

```python
#Task 1 - Coin Change Problem:
def min_coins(coins, amount):

    # Initialize the DP table with a large value (infinity)
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0  # Base case: 0 coins are needed for amount 0

    # Fill the DP table
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1

# Test case
coins = [1, 2, 5]
amount = 11
print(min_coins(coins, amount))  # Output: 3

```

```python
#Task 2 - Longest Common Subsequence (LCS):
def longest_common_subsequence(s1, s2):
    
    n, m = len(s1), len(s2)
    dp = [[0] * (m + 1) for _ in range(n + 1)]

    # Fill the DP table
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[n][m]

# Test case
s1 = "abcde"
s2 = "ace"
print(longest_common_subsequence(s1, s2))
```

```python
#Task 3 - 0/1 Knapsack Problem:
def knapsack(weights, values, capacity):
    
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    # Fill the DP table
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][capacity]

# Test case
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
print(knapsack(weights, values, capacity))
```
